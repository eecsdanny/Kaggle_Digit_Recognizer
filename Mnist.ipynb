```
import pandas as pd

train = pd.read_csv('/content/train.csv')
test = pd.read_csv('/content/test.csv')
```

```
X_train = train.drop('label',axis=1)
Y_train = train['label']
X_train = X_train.values.reshape(-1,28,28,1)
test = test.values.reshape(-1,28,28,1)
```

```
X_train = X_train/255
test = test/255
```

```
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(X_train,Y_train,test_size = 0.1)
y_test = tf.keras.utils.to_categorical(y_test, num_classes = 10)
y_train = tf.keras.utils.to_categorical(y_train, num_classes = 10)
```

```
import tensorflow as tf
import tensorflow.keras.layers as layers
model_input = layers.Input((28,28,1))

L =layers.Conv2D(64,kernel_size = (6,6),strides=(2,2), padding='same', activation='relu')(model_input)
L =layers.BatchNormalization()(L)
L =layers.Dropout(0.3)(L)


L =layers.Conv2D(64,kernel_size = (5,5),strides=(1,1), padding='same', activation='relu')(L)
L =layers.BatchNormalization()(L)
L =layers.Dropout(0.3)(L)


L =layers.Conv2D(32,kernel_size = (3,3),strides=(1,1), padding='same', activation='relu')(L)
L =layers.MaxPool2D(pool_size=(2,2))(L)
L =layers.BatchNormalization()(L)
L =layers.Dropout(0.3)(L)

L =layers.Conv2D(32,kernel_size = (3,3),strides=(1,1), padding='same', activation='relu')(L)
L =layers.BatchNormalization()(L)
L =layers.Dropout(0.3)(L)


L =layers.Conv2D(32,kernel_size = (2,2),strides=(1,1), padding='same', activation='relu')(L)
L =layers.BatchNormalization()(L)
L =layers.Dropout(0.3)(L)


L =layers.Conv2D(32,kernel_size = (2,2),strides=(1,1), padding='same', activation='relu')(L)
L =layers.BatchNormalization()(L)
L =layers.MaxPool2D(pool_size=(2,2))(L)
L =layers.BatchNormalization()(L)

L =layers.Flatten()(L)
L =layers.Dense(512,activation ='relu')(L)

L =layers.Dropout(0.5)(L)

out =layers.Dense(10,activation ='softmax')(L)
model = tf.keras.models.Model(model_input,out)

from tensorflow.keras.optimizers import Adam
opt = Adam(lr=0.001)

model.compile(optimizer=opt, loss = "categorical_crossentropy", metrics=["accuracy"])
annealer = tf.keras.callbacks.LearningRateScheduler(lambda x: 1e-3 * 0.95 ** x)
```

```
datagen = tf.keras.preprocessing.image.ImageDataGenerator(rotation_range=10,width_shift_range=0.12,height_shift_range=0.12)
history = model.fit_generator( datagen.flow(x_train,y_train, batch_size=128),epochs=200,validation_data = (x_test,y_test),verbose = 2)
# history = model.fit_generator( datagen.flow(x_train,y_train, batch_size=128),epochs=120,validation_data = (x_test,y_test),verbose = 2,callbacks=[annealer])
```
